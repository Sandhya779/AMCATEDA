```python
def UniVariateAnalysis(numerical_data):
    for column in numerical_data:
        print()
        print('$' * 5,' ' * 5, column.upper(),' ' * 5, '$' * 5)
        print('Mean:',numerical_data[column].mean())
        print('Median:',numerical_data[column].median())
        print('Mode:',numerical_data[column].mode())
        print('Minimum:',numerical_data[column].min())
        print('Maximum:',numerical_data[column].max())
        print('Kurtosis:',numerical_data[column].kurt())
        print()
        print("_" * 40)
        print()
UniVariateAnalysis(df[['Salary', '10percentage', '12percentage', 'collegeGPA', 'English',
       'Logical', 'Quant', 'Domain', 'ComputerProgramming',
       'ElectronicsAndSemicon', 'ComputerScience', 'MechanicalEngg',
       'ElectricalEngg', 'TelecomEngg', 'CivilEngg', 'conscientiousness',
       'agreeableness', 'extraversion', 'nueroticism',
       'openess_to_experience']])
```

# Research Question 1
# Times of India article dated Jan 18, 2019 states that "After doing your Computer Science Engineering if you take up jobs as a programming Analyst,Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate

```python
from scipy import stats
relevant_roles=['programmer Analyst','software engineer','hardware engineer','associate engineer']
filtered_df=df[df['Designation'].str.contains('|'.join(relevant_roles),case=False,na=False)]

salary_data=filtered_df['Salary']
claimed_mean_salary = 2.75 * 100000
t_stat,p_value=stats.ttest_1samp(salary_data,claimed_mean_salary)
print(f"Mean Salary of Selected Roles: {salary_data.mean():.2f}")
print(f"Claimed Mean Salary: {claimed_mean_salary:.2f}")
```

# Research question 2
# Relationship between gender and specialization

```python
from scipy.stats import chi2_contingency
gender_spec=pd.crosstab(df['Gender'],df['Specialization'])
chi2_value,p_value,dof,expected=chi2_contingency(gender_spec)
print(f"Chi-square test p_value for gender and specialisation:{p_value}")
```

## Insights
- The Chi- square test results showed a statistically significant
- There is a strong relationship between gender and specialization.
- Therefore the choice of specialization is dependent on gender in our dataset
